# FEATURE EXTRACTION

```python
# Converting object data type to datetime in Journey_date column 
df_fp['Journey_date'] =  pd.to_datetime(df_fp['Journey_date'])
```

```python
# Extracting Journey year,month and day from Journey_date

# Extracting year
df_fp["Journey_year"]=pd.to_datetime(df_fp.Journey_date, format="%Y/%m/%d").dt.year

# Extracting month
df_fp["Journey_mon"]=pd.to_datetime(df_fp.Journey_date, format="%Y/%m/%d").dt.month

# Extracting day
df_fp["Journey_day"]=pd.to_datetime(df_fp.Journey_date, format="%Y/%m/%d").dt.day
```

```python
# Checking value count of Journey_year column
df_fp.Journey_year.value_counts()
```

Since all the entries in Journey_year column are same, let's drop as it will not help in our analysis.

```python
# Dropping Journey_year column
df_fp = df_fp.drop(["Journey_year"],axis=1)
```

```python
# Checking valuecount of Journey_mon column
df_fp.Journey_mon.value_counts()
```

```python
# Dropping Journey_mon column
df_fp = df_fp.drop(["Journey_mon"],axis=1)
```

```python
# Dropping Journey_date column
df_fp = df_fp.drop(["Journey_date"],axis=1)
```

```python
# Converting object data type to datetime 
df_fp['Dtime'] =  pd.to_datetime(df_fp['Dtime'])
df_fp['Atime'] =  pd.to_datetime(df_fp['Atime'])
```

```python
# Checking the data types of all columns again
df_fp.dtypes
```

# BEST RANDOM RATE

```python
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
maxAccu=0
maxRS=0
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.30, random_state = i)
    KNN=neighbors.KNeighborsRegressor()
    KNN.fit(x_train,y_train)
    pred_rs=KNN.predict(x_test)
    r2=r2_score(y_test,pred_rs)
    if r2>maxAccu:
        maxAccu=r2
        maxRS=i
print("Best R2 Score is :",maxAccu," on Random State :",maxRS)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRS)
```

# LINEAR REGRESSION

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python

from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)

pred_lr=lr.predict(x_test)
print('R2_Score: ',r2_score(y_test,pred_lr))
print('Mean absolute error: ',mean_absolute_error(y_test,pred_lr))
print('Mean squared error: ',mean_squared_error(y_test,pred_lr))
print('Root Mean squared error: ',np.sqrt(mean_squared_error(y_test,pred_lr)))
```

# DECISION TREE REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr.score(x_train,y_train)

pred_dtr=dtr.predict(x_test)
print('R2_Score: ',r2_score(y_test,pred_dtr))
print('Mean absolute error: ',mean_absolute_error(y_test,pred_dtr))
print('Mean squared error: ',mean_squared_error(y_test,pred_dtr))
print('Root Mean squared error: ',np.sqrt(mean_squared_error(y_test,pred_dtr)))
```

# KNN

```python
from sklearn import neighbors

knn = neighbors.KNeighborsRegressor()
knn.fit(x_train,y_train)
knn.score(x_train,y_train)

pred_knn=knn.predict(x_test)
print('R2_Score: ',r2_score(y_test,pred_knn))
print('Mean absolute error: ',mean_absolute_error(y_test,pred_knn))
print('Mean squared error: ',mean_squared_error(y_test,pred_knn))
print('Root Mean squared error: ',np.sqrt(mean_squared_error(y_test,pred_knn)))
```
